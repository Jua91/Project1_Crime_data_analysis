```python
# import full table csv and read in
data_df = pd.read_csv("../../Data_files/clean_data/states_clean_data/Crime_Data_Analysis(combined_info).csv")
data_df.head()
```

# Map

```python
data_df["Color Number"]= ""
data_df.head()
```

```python
# set condition to assign float for new column
# red state
data_df.loc[data_df["Affiliated Party Color"] == "Red","Color Number"] = 1
# purple state
data_df.loc[data_df["Affiliated Party Color"] == "Purple", "Color Number"] = 2
# blue state
data_df.loc[data_df["Affiliated Party Color"] == "Blue","Color Number"] = 3
data_df.head()
```

```python
for col in data_df.columns:
    data_df[col] = data_df[col].astype(str)
data_df['text'] = data_df['State'] + " " + "|" + ' Population:'  + " "+ data_df['Population'] +'<br>' + \
    'VCR: '  + data_df['Violent Crime per 100,000 inhabitants'] + " " + "|" + ' PCR: '  + data_df['Property Crime per 100,000 inhabitants'] + '<br>' + \
    'HCR: '  + data_df['Hate Crime per 100,000 inhabitants'] + '<br>' + \
    'UE Rate' + " " + data_df['Unemployment (UE) Rate'] + " " + "|" + ' UE Rank ' + data_df['UE Rank among States']


fig = go.Figure(data=go.Choropleth(
    locations=data_df['State Abbrev.'],
    z = data_df['Color Number'],  #replace with party affiliation
    locationmode='USA-states',
    colorscale=[[0, 'red'], [0.3, 'purple'], [1.0, 'rgb(0, 0, 255)']], # edit for state party color
    autocolorscale=False,
    text=data_df['text'], # hover text
    marker_line_color='white', # to separate steps
    colorbar_title="Voting Affiliation Scale"
))

fig.update_layout(
    title_text='US Crime Data Analysis',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True, # lakes
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```

# Scatter Plots and Regression Analysis

```python
# Assign the variables for red states
red_violent_crime = data_df.loc[data_df["Affiliated Party Color"]=="Red","Violent Crime per 100,000 inhabitants"]
red_property_crime = data_df.loc[data_df["Affiliated Party Color"]=="Red","Property Crime per 100,000 inhabitants"]
red_hate_crime=data_df.loc[data_df["Affiliated Party Color"]=="Red","Hate Crime per 100,000 inhabitants"]
red_unemployment=data_df.loc[data_df["Affiliated Party Color"]=="Red","Unemployment (UE) Rate"]

# Assign the variables for purple states
purple_violent_crime = data_df.loc[data_df["Affiliated Party Color"]=="Purple","Violent Crime per 100,000 inhabitants"]
purple_property_crime = data_df.loc[data_df["Affiliated Party Color"]=="Purple","Property Crime per 100,000 inhabitants"]
purple_hate_crime=data_df.loc[data_df["Affiliated Party Color"]=="Purple","Hate Crime per 100,000 inhabitants"]
purple_unemployment=data_df.loc[data_df["Affiliated Party Color"]=="Purple","Unemployment (UE) Rate"]

# Assign the variables for blue states
blue_violent_crime= data_df.loc[data_df["Affiliated Party Color"]=="Blue","Violent Crime per 100,000 inhabitants"]
blue_property_crime = data_df.loc[data_df["Affiliated Party Color"]=="Blue","Property Crime per 100,000 inhabitants"]
blue_hate_crime= data_df.loc[data_df["Affiliated Party Color"]=="Blue","Hate Crime per 100,000 inhabitants"]
blue_unemployment= data_df.loc[data_df["Affiliated Party Color"]=="Blue","Unemployment (UE) Rate"]
```

```python
def regress_scatter(x,y,a,b,c,d,e,f):

    (slope,intercept,rvalue,pvalue,stderr)=st.linregress(x,y) # Calculate the linear regression
    regress=slope*x+intercept # Calculate the new y values using the linear equation
    lin_equation=f"y={round(slope,3)}*x+{round(intercept,2)}"
    correlation=st.pearsonr(x,y)

    #Display the line plot
    plt.figure(figsize=(10,6))
    plt.plot(x,regress,color="black")
    plt.annotate(lin_equation,xy_coord,fontsize="xx-large",color="black",fontweight="semibold")
    plt.annotate(f"Correlation: {round(correlation[0],3)}\nR-squared: {round(rvalue**2,3)}\nP-value: {round(pvalue,3)}",\
                coord,fontsize="large",color="black")

    #Display the scatter plots with different colors
    plt.scatter(x=a,y=b, c="red", label="Red States")
    plt.scatter(x=c,y=d, c="blue", label="Blue States")
    plt.scatter(x=e,y=f, c="purple", label="Purple States")
    plt.legend()
```

```python
#Assign x and y values for regression
x=data_df["Property Crime per 100,000 inhabitants"]
y=data_df["Violent Crime per 100,000 inhabitants"]

xy_coord=(2700,700)
coord=(2000,700)

regress_scatter(x,y,red_property_crime,red_violent_crime,blue_property_crime,blue_violent_crime,\
               purple_property_crime,purple_violent_crime)

# Set the title and labels
plt.title("Property Crime vs Violent Crime per 100,000 inhabitants")
plt.xlabel("Property Crime per 100,000 inhabitants")
plt.ylabel("Violent Crime per 100,000 inhabitants")

plt.savefig("../../Figures/Scatter-property_vs_violent_per_100k.png")
plt.show()
```

```python
#Assign x and y values for regression
x=data_df["Hate Crime per 100,000 inhabitants"]
y=data_df["Violent Crime per 100,000 inhabitants"]

xy_coord=(5,600)
coord=(4,750)

regress_scatter(x,y,red_hate_crime,red_violent_crime,blue_hate_crime,blue_violent_crime,\
               purple_hate_crime,purple_violent_crime)

# Set the title and labels
plt.title("Hate Crime vs Violent Crime per 100,000 inhabitants")
plt.xlabel("Hate Crime per 100,000 inhabitants")
plt.ylabel("Violent Crime per 100,000 inhabitants")

# Save chart
plt.savefig("../../Figures/Scatter-hate_vs_violent_per_100k.png")
plt.show()
```
